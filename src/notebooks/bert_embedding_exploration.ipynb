```python
# visualize embedding clusters (PCA)
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
coords = pca.fit_transform(emb)
plt.scatter(coords[:,0], coords[:,1], s=8)
plt.title("BERT Embeddings (PCA)")
plt.show()
```

```python
# visualize embedding clusters (t-SNE)
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, perplexity=30, learning_rate=200)
coords_tsne = tsne.fit_transform(emb)
plt.scatter(coords_tsne[:,0], coords_tsne[:,1], s=8)
plt.title("BERT Embeddings (t-SNE)")
plt.show()
```
